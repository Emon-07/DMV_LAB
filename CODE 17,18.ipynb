```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate synthetic data with clusters, outliers, and correlation
np.random.seed(42)

# Cluster 1 (Positive correlation)
cluster1_x = np.random.normal(5, 1, 100)
cluster1_y = 2 * cluster1_x + np.random.normal(0, 0.5, 100) + 5

# Cluster 2 (Negative correlation, slightly offset)
cluster2_x = np.random.normal(15, 1.5, 80)
cluster2_y = -1.5 * cluster2_x + np.random.normal(0, 0.8, 80) + 30

# Cluster 3 (Another positive correlation cluster, denser)
cluster3_x = np.random.normal(10, 0.5, 60)
cluster3_y = 1.0 * cluster3_x + np.random.normal(0, 0.3, 60) - 2

# Outliers
outliers_x = [2, 18, 8, 12, 10]
outliers_y = [20, 5, 25, 0, 15]

# Combine all data
x = np.concatenate([cluster1_x, cluster2_x, cluster3_x, outliers_x])
y = np.concatenate([cluster1_y, cluster2_y, cluster3_y, outliers_y])

# Create a DataFrame
sample_df = pd.DataFrame({'Feature_X': x, 'Feature_Y': y})

# Plotting the scatter plot
plt.figure(figsize=(10, 7))
sns.scatterplot(x='Feature_X', y='Feature_Y', data=sample_df, alpha=0.7)
plt.title('Scatter Plot with Clusters, Outliers, and Correlation')
plt.xlabel('Feature X')
plt.ylabel('Feature Y')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
